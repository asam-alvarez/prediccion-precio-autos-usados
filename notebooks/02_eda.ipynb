## **Análisis exploratorio de datos (EDA)**
* Estadísticas descripticas (precio, kilometraje, potencia, etc.).
* Distribución de variables clave (precio, año, tipo de combustible, etc.).
* Correlaciones.
* Visualizaciones

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_data_cleaned_data = "../data/processed/cleaned_data.csv"

df = pd.read_csv(path_data_cleaned_data)

df
```

```python
# Estadísticas descriptivas
print("\nEstadísticas descriptivas de variables numéricas")
display(df.describe())

print("\nEstadísticas descriptivas de variables categóricas")
display(df.describe(include="object"))
```

```python
# Gráfico de pares de variables numéricas para explorar posibles relaciones
num_cols = df.select_dtypes("number")
sns.pairplot(num_cols)
plt.show()
```

```python
# Analisis de la distribución de precios
plt.figure(figsize=(10, 10))
sns.histplot(x=df["price_in_euro"], bins=50, kde=True)
plt.title("Distribución de precios")
plt.xlabel("Precio (Euros)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Diagrama de caja de los precios
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["price_in_euro"])
plt.title("Diagrama de caja de precios de autos")
plt.show()
```

```python
# Análisis de correlaciones
plt.figure(figsize=(10, 10))
corr = df.select_dtypes("number").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor de correlaciones entre variables numéricas")
plt.show()
```
